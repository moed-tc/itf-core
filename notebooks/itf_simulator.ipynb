## 4. Visualização Numérica – Autovalores do TC

Vamos parametrizar as funções f_i de forma simples para simulação numérica.

```python
# Parametrização exemplo (pode ser aprendida por PINN depois)
T1_val, T2_val, T3_val, T4_val = sp.symbols('T1 T2 T3 T4')

# Exemplo realista: compressão isotrópica + pequena anisotropia
f1_sub = 1 + 0.1 * T1_val**2 + 0.05 * (T2_val - T3_val)**2
f2_sub = 1 + 0.08 * T2_val**2
f3_sub = 1 + 0.12 * T3_val**2
f4_sub = 1 + 0.06 * T4_val**2

TC_num = TC.subs({f1: f1_sub, f2: f2_sub, f3: f3_sub, f4: f4_sub})

# Função para calcular autovalores numericamente
TC_lamb = sp.lambdify((T1_val, T2_val, T3_val, T4_val), TC_num.eigenvals(), 'numpy')

# Variação ao longo de uma trajetória (ex: compressão crescente isotrópica)
t = np.linspace(0, 2.0, 200)
eig1, eig2, eig3, eig4 = [], [], [], []

for ti in t:
    vals = TC_lamb(ti, ti*0.95, ti*1.05, ti*0.9)  # pequena anisotropia
    eig_list = sorted(list(vals.keys()), key=float)  # SymPy retorna dict
    eig1.append(float(eig_list[0]))
    eig2.append(float(eig_list[1]))
    eig3.append(float(eig_list[2]))
    eig4.append(float(eig_list[3]))

plt.figure(figsize=(12, 7))
plt.plot(t, eig1, label=r'$\lambda_1$', linewidth=2)
plt.plot(t, eig2, label=r'$\lambda_2$', linewidth=2)
plt.plot(t, eig3, label=r'$\lambda_3$', linewidth=2)
plt.plot(t, eig4, label=r'$\lambda_4$', linewidth=2)
plt.title('Autovalores do Tensor de Compressão TC durante evolução', fontsize=16)
plt.xlabel('Parâmetro de compressão média (τ)', fontsize=14)
plt.ylabel('Autovalores λ_i', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.show()

print("Interpretação:\n- Autovalores próximos → compressão isotrópica (energia escura efetiva)\n- Dispersão → anisotropia (matéria escura efetiva)\n- Crescimento forte → bounce via IT")
```
